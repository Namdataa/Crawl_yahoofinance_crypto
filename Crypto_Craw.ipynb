```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.common.exceptions import ElementNotInteractableException, NoSuchElementException

driver = webdriver.Chrome()
driver.get('https://finance.yahoo.com/crypto/?count=100&offset=0')
time.sleep(3)
heads = []
values = []

# Thu thp d liu t trang ban u
name = driver.find_elements("xpath", '//*[@id="scr-res-table"]/div[1]/table/thead/tr')
for i in name:
    heads.append(i.text)

while True:
    time.sleep(3)
    
    # Thu thp d liu t mi dng trn trang
    for n in range(1, 101):
        try:
            volumn = driver.find_elements("xpath", f'//*[@id="scr-res-table"]/div[1]/table/tbody/tr[{n}]')
            for j in volumn:
                values.append(j.text)
                print(j.text)
        except:
            pass
    
    # Kim tra nt "Next" c kh dng khng
    try:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="scr-res-table"]/div[2]/button[3]')))
        page = driver.find_element("xpath", '//*[@id="scr-res-table"]/div[2]/button[3]')
        page.click()
    except (ElementNotInteractableException, NoSuchElementException):
        print("QQ")
        break

# In ra kt qu thu thp c
print(heads)
print(values)
```

```python
list_chu = []
list_so = []
list_truoc = []
list_sau = []
for x in values:
    # Tm v tr ca du '-'
    x=x.replace(',', '')
    index_backslash = x.rfind('\n')

# Nu khng tm thy '\n' th khng lm g c
    if index_backslash != -1:
    # Tip tc tm v tr ca '-' t v tr '\n' tm c
        index_dash = x.rfind('-', 0, index_backslash)

    # Nu khng tm thy '-' th khng lm g c
    if index_dash != -1:
        # Ly key t v tr du '-' n v tr du '\n'
        key_reversed = x[index_dash + 1:index_backslash]
    #print(key[::-1])
    index_key = x.rfind(key_reversed)    # Tch chui t bn phi n key
    list_truoc.append(x[:index_key] + key_reversed )#+f'{key[::-1]}')
    list_sau.append(x[index_key + len(key_reversed):])
#for chuoi in list_sau:
   # index_dau_tien = None
  #  for i, char in enumerate(chuoi):
  #      if char.isdigit() :
 #           index_dau_tien = i
#            break
  #  if index_dau_tien is not None:
 #       list_chu.append(chuoi[:index_dau_tien].strip())
  #      list_so.append(chuoi[index_dau_tien:].strip())
 #   else:
#        print("Khng tm thy k t s trong chui.")
list_sau = [s.replace(',', '').replace(' ', ', ') for s in list_sau]
for i in range(len(list_truoc)):
    list_truoc[i] = list_truoc[i].replace('\n', ', ')
chuoi_tong_hop = []
for dau, sau in zip(list_truoc, list_sau):
   chuoi_tong_hop.append(dau + sau)

chuoi_tong_hop
```

```python
data = [item.split(", ") for item in chuoi_tong_hop]

# m bo mi phn t trong d liu c  12 gi tr
for item in data:
    while len(item) < len(heads):
        item.append(None)

# To DataFrame t d liu  chia v tn ct
df = pd.DataFrame(data, columns=heads)

# In DataFrame
df.head(50)
excel_file = "Crypto_Crawl.xlsx"

# Xut DataFrame ra Excel
df.to_excel(excel_file, index=False)
```
